# Categorize for receipe category, cooking methods and ingredient

```python
import os
import pandas as pd

base_path = './first_clean_recipe/'
files = os.listdir(base_path)
```

## Receipe category

```python
# categories list
Categories = ['汤', '羹', '粥', '膏', '面', '粉', '馄饨', '馒头', '饼', '肉', '煎', '茶', '浆', '油']
# result list
categories = pd.DataFrame(columns=['category', 'Food_Name'])

for file_name in files:
    df = pd.read_csv(os.path.join(base_path, file_name))
    
    for foodname in df['Food_Name']:
        ctgr = 0
        for Category in Categories:
            if Category in foodname:
                food_contain = pd.DataFrame({'category': Category, 'Food_Name': foodname}, index=pd.RangeIndex(start=0, stop=len(Category)))
                categories = pd.concat([categories, food_contain], ignore_index=True)
                ctgr = 1 # categorized as (at least) one of the category list
        if ctgr == 0:
            food_contain = pd.DataFrame({'category': '菜品', 'Food_Name': foodname}, index=pd.RangeIndex(start=0, stop=len(Category)))
            categories = pd.concat([categories, food_contain], ignore_index=True)
        
categories
```

```python
categories.to_csv(f'./categorize/categories.csv', index=False, encoding='utf-8-sig')
```

## Methods

```python
# mothods list
Methods = ['煮', '蒸', '熬', '煎', '焙', '淹', '炙', '炒', '炸', '烧']
# result list
methods = pd.DataFrame(columns=['method', 'Food_Name'])

for file_name in files:
    df = pd.read_csv(os.path.join(base_path, file_name), encoding='utf-8')
    
    for index, row in df.iterrows():
        if pd.notnull(row['Steps']):
            for Method in Methods:
                if Method in row['Steps']:
                    food_contain = pd.DataFrame({'method': Method, 'Food_Name': row['Food_Name']}, index=pd.RangeIndex(start=0, stop=len(Category)))
                    methods = pd.concat([methods, food_contain], ignore_index=True)
        
methods
```

```python
ingredient_df = pd.read_csv(f'unique_categorize/unique_ingredient.csv',encoding = "utf-8-sig")
category_list={'dairy_product':['酥油','奶'],
               'meat':['鸡','猪','羊','驴','肉','鸽','鸭','牛','鹿','兔','马','蹄','肠','雁'],
               'vegetable':['菜','粟','葱','萝卜','笋','黄瓜','茄','秦椒','韭','藕'],
               'plant':['草','花','叶','豆','萸','乌头','荆芥穗','桂'],
               'fruit':['石榴','桃','橘','枣','梅','杏','栀'],
               'seafood':['鱼'],
               'condiment':['盐','糖'],
               'spice':['蜀椒','姜','蒜','椒'],
               'juice':['汁'],
               'grain':['米','饼','面','大麦','仁'],
               'tea':['茶'],
               'chinese_medicinal_material':['夏','角','枸杞','香','皮','仁','荆实','人参','砂','膏','官桂','味']}
for ingredient in ingredient_df['Ingredient'].values:
    flag=True
    for category in category_list:
        for category_ingredient in category_list[category]:
            if category_ingredient in ingredient:
                ingredient_df.loc[ingredient_df['Ingredient']==ingredient,'category']=category
                flag=False
                break
    if flag:
        ingredient_df.loc[ingredient_df['Ingredient']==ingredient,'category']='other'

display(ingredient_df.head(10))
        
```
