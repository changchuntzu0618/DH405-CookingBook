```python
import os
import pandas as pd

base_path = './first_clean_recipe/'
files = os.listdir(base_path)
```

## Category

```python
# categories list
Categories = ['汤', '羹', '粥', '膏', '面', '粉', '馄饨', '馒头', '饼', '肉', '煎', '茶', '浆', '油']
# result list
categories = pd.DataFrame(columns=['category', 'Food_Name'])

for file_name in files:
    df = pd.read_csv(os.path.join(base_path, file_name))
    
    for foodname in df['Food_Name']:
        ctgr = 0
        for Category in Categories:
            if Category in foodname:
                food_contain = pd.DataFrame({'category': Category, 'Food_Name': foodname}, index=pd.RangeIndex(start=0, stop=len(Category)))
                categories = pd.concat([categories, food_contain], ignore_index=True)
                ctgr = 1 # categorized as (at least) one of the category list
        if ctgr == 0:
            food_contain = pd.DataFrame({'category': '菜品', 'Food_Name': foodname}, index=pd.RangeIndex(start=0, stop=len(Category)))
            categories = pd.concat([categories, food_contain], ignore_index=True)
        
categories
```

```python
categories.to_csv(f'./categories.csv', index=False, encoding='utf-8-sig')
```
